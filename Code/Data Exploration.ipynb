# 6. Transforming variables test

```python
df = pd.read_csv('../Data/data.csv')
```

```python
import numpy as np
```

```python
np.power(df['Age'],1/3).plot.hist()
```

```python
bins = [0,18,80]
```

```python
group = ['Children','Adults']
```
